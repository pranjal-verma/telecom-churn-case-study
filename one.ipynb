```python
cols_with_highnull_values = []
for col in data_subset.columns:
    if data_subset[col].isnull().sum() * 100 / len(data_subset) > 50:
        cols_with_highnull_values.append(col)
print(cols_with_highnull_values)
```

```python
#drop columns with high null values
# count_rech_3g_8, av_rech_amt_data_8: recharge count for 3g is an important feature, so we will not drop it, we can replace null values with 0
# arpu_2g_6,arpu_3g_8, arpu_3g_6: Average Revenue these are important features, so we will not drop them, we can replace null values with average of the column since this is an average 
# fb_user_6, date_of_last_rech_data_6,
replace_null_with_mean = ['arpu_2g_6', 'arpu_3g_8', 'arpu_3g_6', 'av_rech_amt_data_8', 'arpu_2g_8','av_rech_amt_data_6','av_rech_amt_data_7']
replace_null_with_zero = ['count_rech_3g_8', 'night_pck_user_8',
                        'count_rech_3g_6', 'total_rech_data_6','max_rech_data_8', 'max_rech_data_7']
columnn_to_drop = ['fb_user_6', 'date_of_last_rech_data_6', 'date_of_last_rech_6', 'date_of_last_rech_7', 'date_of_last_rech_8']

sum([len(replace_null_with_mean)
,len(replace_null_with_zero)
,len(columnn_to_drop)])
```

```python
len(cols_with_highnull_values)
```

```python
# replace null values with mean
for col in replace_null_with_mean:
    data_subset[col].fillna(data_subset[col].mean(), inplace=True)

# replace null values with zero
for col in replace_null_with_zero:
    data_subset[col].fillna(0, inplace=True)

# drop columns
for col in columnn_to_drop:
    data_subset.drop(col, axis=1, inplace=True)

data_subset.isnull().sum() * 100 / len(data_subset)
```
